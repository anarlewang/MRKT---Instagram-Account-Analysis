### 1. Compute Cumulative Number of Posts as of Post Created (n_cumpost)

```python
# sort by username and post creation
df = data.sort_values(by=['User Name','Post Created'], ascending=True)
df = df.reset_index(drop=True)

# count from 1 for each username
df['n_cumpost'] = df.groupby(['User Name']).cumcount() + 1
df.loc[df['User Name'] == 'chelseaspack']
```

### 2. Number of Posts in Past 6/3/1 Months (n_postsP6M, n_postsP3M, n_postsP1M)

### 3. Compute Number of Days Since Last Posted (d_lastpost)

```python
# indicate 1 for first post
df['first_post'] = np.where(df['n_cumpost']==1, 1, 0)
```

```python
# compute date diff from last post
df['d_lastpost'] = df.groupby('User Name')['Post Created'].diff().apply(lambda x: x.days)
df.loc[df['User Name'] == 'chelseaspack']
```

### 4. Compute Number of Days since First posted (d_firstpost)

```python
# take days since last posted
df['d_firstpost'] = df['d_lastpost']

# accumulate days since last posted for each username
df.loc[:, 'd_firstpost'] = df.groupby(['User Name', 'first_post'])['d_firstpost'].cumsum()
df.loc[df['User Name'] == 'chelseaspack']
```
